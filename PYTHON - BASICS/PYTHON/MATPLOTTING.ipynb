### PLOTTING


```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([1,2])
ypoints=np.array([1,2])

plt.plot(xpoints,ypoints)
plt.show()
```

### MARKERS

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([5,3,3,2])

plt.plot(ypoints,marker = 'o')
plt.show()
```

### LINES

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([5,3,3,2])

plt.plot(ypoints,linestyle = "none")
plt.show()
```

### LABELS

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,34,5,5])
y=np.array([5,3,3,2])

plt.plot(x,y)

plt.xlabel("no of students")
plt.ylabel("no idea")
plt.show()
```

### PLOTTING WITH LABELS AND TITLE

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,34,5,5])
y=np.array([5,3,3,2])

plt.plot(x,y)
plt.title("anvika")
plt.xlabel("no of students")
plt.ylabel("no idea")
plt.grid()
plt.show()
```

### SCATTER

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,34,5,5])
y=np.array([5,3,3,2])
plt.scatter(x,y)
plt.show()
```

### BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,34,5,5])
y=np.array([5,3,3,2])
plt.bar(x,y)
plt.barh(x,y)
plt.show()
```

### HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal([2,34,5,5])
plt.hist(x)
plt.show
```

### PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([5,3,3,2])
mylabels=["apple","banana","gauva","pazham"]
plt.pie(y,labels=mylabels)
plt.show()
```
